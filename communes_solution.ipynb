B.5 Compute the population density at the canton level and rank the cantons from most dense to least dense.

```python
population = df_Clean.groupby("Canton")["Residents"].sum()
surface = df_Clean.groupby("Canton")["Surface area in km²"].sum()
df4 = pd.concat([population,surface], axis=1)
df4["Density"] = (df4["Residents"]/df4["Surface area in km²"])
df4.sort_values('Density', ascending=False)
```

We note that the density of the 1st canton is more than twice as high as that of the second and 3rd cantons, and just under half that of the second.

From the 4th canton onwards, density decreases more slowly.

B.7 Compute the difference between the maximum and minimum elevations for each canton.
Find the top 5 cantons that have the largest range of elevations?

```python
haut = df_Clean.groupby("Canton")["Elevation"].max()
haut.rename("Max", inplace = True)
bas = df_Clean.groupby("Canton")["Elevation"].min()
bas.rename("Min", inplace = True)         
df7 = pd.concat([bas,haut], axis=1)
df7["Delta"] = (df7["Max"]-df7["Min"])
df7.sort_values('Delta', ascending=False, inplace = True)
df7[:5]
```

C.3

```python
young = df_Clean["0-19 years"].tolist()
tranches = ((5,10),
            (10,15),
            (15,20),
            (20,25),
            (25,30),
            (30,35))
pourcentage = {}
for x in young:
    
    for t in tranches:
        if t[0] <= x < t[1]:
            pourcentage[t] = pourcentage.get(t, 0) +1
            break
pourcentage           
```

```python
listeValues = []
indice = 0
for k in sorted(pourcentage):
    for i in range(pourcentage[k]):
        listeValues.append(indice)
    indice += 1    
values = np.array(listeValues)
```

```python
labels = [f"{t[0]}%-{t[1]}%" for t in tranches]
y_cor = np.arange(len(labels))

plt.figure(figsize=(5,4))
plt.hist(values, edgecolor="red", orientation='horizontal', histtype = "bar", range = (0,5), bins = 5, align = "left")
plt.xlabel('pourcents of young residents (0-19yrs)')
plt.ylabel('numbers of Communes')
plt.title('pourcents of young residents (0-19yrs) in communes')
plt.yticks(y_cor,labels)

plt.show()
```
